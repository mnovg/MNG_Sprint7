Calculadora de l'índex de massa corporal.

Escriu una funció que calculi l'IMC ingressat per l'usuari/ària, és a dir, qui ho executi haurà d'ingressar aquestes dades.  
La funció ha de classificar el resultat en les seves respectives categories.

```python
# Función que calcula el IMC
def calculoIMC (peso,altura):
    IMC = round(peso/(altura**2),2)
    print("Su IMC es: ",IMC)
    if IMC<18.5:
        print("\nSegún su IMC, su situación corporal es de BAJO PESO")
    elif IMC<24.9:
        print("\nSegún su IMC, su situación corporal es de  PESO NORMAL")
    elif IMC<29.9:
        print("\nSegún su IMC, su situación corporal es de  SOBREPESO")
    else:
        print("\nSegún su IMC, su situación corporal es de  OBESIDAD")

# Título del cálculo
print("\tCALCULO DEL INDICE DE MASA CORPORAL (IMC)\n")

# Control en la introducción de datos por el promt, para el peso y la altura
pesoIntroducido=True
while pesoIntroducido==True:
  try:
    peso=float(input("Introduzca su peso en kg"))
    print("PESO: ",peso, "kg")
    pesoIntroducido=False
  except:
    print("\nDato erróneo. Por favor intentalo de nuevo (recuerda introducir tu peso en kg)")

alturaIntroducida=True
while alturaIntroducida==True:
  try:
    altura=float(input("Introduzca su altura en m"))
    print("ALTURA: ",altura, "m\n")
    alturaIntroducida=False
  except:
    print("\nDato erróneo. Por favor intentalo de nuevo (recuerda introducir tu altura en m, con los decimales separados por un punto)")

# Llamada a la función que calcula el IMC, pasándole como parámetros el peso y la altura solicitados al usuario
calculoIMC (peso,altura)
```

#  <font color="violet">EXERCICI 1.2</font>
Convertidor de temperatures.

Existeixen diverses unitats de temperatura utilitzades en diferents contextos i regions. Les més comunes són Celsius (°C), Fahrenheit (°F) i Kelvin (K). També existeixen altres unitats com Rankine (°Ra) i Réaumur (°Re). Selecciona almenys 2 conversors, de tal manera que en introduir una temperatura retorni, com a mínim, dues conversions.

```python
def conversorTemperatura():
    # Título del cálculo
    print("\tCONVERSOR DE TEMPERATURA\n")

    # Control en la introducción de datos por el promt, para la temperatura
    temperaturaIntroducida=True
    while temperaturaIntroducida==True:
        try:
            temperatura=float(input("Introduzca la temperatura en Celsius"))
            print("TEMPERATURA: ",temperatura, "ºC\n")
            temperaturaIntroducida=False
        except:
            print("\nDato erróneo. Por favor intentalo de nuevo (recuerda introducir la temperatura en Celsius (ºC), con los decimales separados por un punto)")

    # Menú de selección del conversor
    print("SELECCION CONVERSOR TEMPERATURA")
    print("1: Celsius (ºC) a Fahrenheit (ºF)\n" \
        "2: Celsius (ºC) a Kelvin (ºK)\n" \
        "3: Celsius (ºC) a Rankine (ºRa)\n" \
        "4: Celsius (ºC) a Réamur (ºRe)\n" \
        "5: Salir")

    # Conversión en función de la opción elegida
    while True:
        opcion=int(input("Selecciona uno de los conversores de temperatura:"))
        print("\nLa opción elegida es: ",opcion)
        if opcion==1:
            fahrenheit=round(((temperatura*1.8)+32),2)
            print(f"{temperatura} ºC son {fahrenheit} ºF")
        elif opcion==2:
            kelvin=round((temperatura+273.15),2)
            print(f"{temperatura} ºC son {kelvin} ºK")
        elif opcion==3:
            rankine=round(((temperatura*1.8)+491.67),2)
            print(f"{temperatura} ºC son {rankine} ºRa")
        elif opcion==4:
            reamur=round(((temperatura*4)/5),2)
            print(f"{temperatura} ºC son {reamur} ºRe")
        elif opcion==5:
            break
        else:
            print("\nOpción inválida. Introduce una opción del 1 al 5.")

# Llamada a la función.
conversorTemperatura()
```

#  <font color="violet">EXERCICI 1.3</font>
Comptador de paraules d'un text.

Escriu una funció que donat un text, mostri les vegades que apareix cada paraula.

```python
def contadorDePalabras():
    # Creación de una lista donde guardar las palabras separadas del texto.
    # Creación del diccionario donde almacenar cada palabla (clave) y el número de veces que se repite en el texto (valor).
    listaPalabras=[]
    diccionarioPalabras={}

    # Eliminar las comas y los puntos de un texto.
    textoIntroducido=input("Introduce el texto que deseas separar en palabras")
    texto=textoIntroducido.replace(",","").replace(".","").replace(";","").replace("!","").replace("?","")

    # Convertir a minúsculas todas las letras del texto, para poder contar la primera despues de un punto.
    textoMinusculas=texto.lower()

    # Copiar las palabras en minúsculas del texto en la lista.
    listaPalabras=textoMinusculas.split()

    # Recorrer lista y traspasar las palabras al diccionario, junto con el número de repeticiones de cada una de las palabras.
    for palabraLista in listaPalabras:
        if palabraLista in diccionarioPalabras:
            diccionarioPalabras[palabraLista]+=1
        else:
            diccionarioPalabras[palabraLista]=1

    # Mostrar el diccionario.
    print("Diccionario:",diccionarioPalabras)

    # Mostrar el resultado detallado por palabras.
    print("\nRepeticiones:")
    for clave,valor in diccionarioPalabras.items():
        print(f"{clave}: {valor}")

# Llamada a la función.
contadorDePalabras()
```

#  <font color="violet">EXERCICI 1.4</font>
Diccionari invers.

Resulta que el client té una enquesta molt antiga que s'emmagatzema en un diccionari i els resultats els necessita al revés, és a dir, intercanviats les claus i els valors. Els valors i claus en el diccionari original són únics; si aquest no és el cas, la funció hauria d'imprimir un missatge d'advertiment.

```python
def diccionarioInverso(diccInverso):
    # Creo el nuevo diccionario donde quedaran las claves y valores del diccionario Inverso correctamente.
    diccDefinitivo={}

    for claveInversa,valorInverso in diccInverso.items():
        if valorInverso not in diccDefinitivo:
            diccDefinitivo[valorInverso]=claveInversa
        else:
            print("Error: multiple keys for one value")
            diccDefinitivo.clear()

    return(diccDefinitivo)

# Creo el diccionario inverso del enunciado sin valores repetidos
diccInversoSinRepeticiones={
    "a":1,
    "b":2,
    "c":3
}

# Creo el diccionario inverso del enunciado con valores repetidos
diccInversoConRepeticiones={
    "x":"apple",
    "y":"banana",
    "z":"banana"
}

# Llamada a la función.
diccionarioInverso(diccInversoConRepeticiones)
diccionarioInverso(diccInversoSinRepeticiones)
```

Diccionari invers amb duplicats

Continuant amb l'exercici 4 del nivell 1: al client es va oblidar de comentar un detall i resulta que els valors en el diccionari original poden duplicar-se i més, per la qual cosa les claus intercanviades poden tenir duplicats. En aquest cas, en l'exercici anterior imprimies un missatge d'advertiment, ara, els valors del diccionari resultant hauran d'emmagatzemar-se com una llista. Tingues en compte que si és un valor únic no ha de ser una llista.

```python
def diccionarioInversoValoresDuplicados(diccInverso):
    diccDefinitivo = {}
    
    for claveInversa, valorInverso in diccInverso.items():
        if valorInverso not in diccDefinitivo:                                                  # Si el valorInverso no existe todavía en el diccionario, lo añado y el su valor claveInversa.
            diccDefinitivo[valorInverso] = claveInversa
        else:                                                                                   # Si el valorInverso ya existe, compruebo si es una lista.
            if isinstance(diccDefinitivo[valorInverso], list):                                  # Con la función isinstance compruebo si el valor guardado con la clave valorInverso es una lista.
                diccDefinitivo[valorInverso].append(claveInversa)                               # Añado al final de la lista el nuevo valor, al estar invertido es lo que se pasa como clave.
            else:                                                                               # Si no es una lista, la creo.
                diccDefinitivo[valorInverso] = [diccDefinitivo[valorInverso], claveInversa]     # Convierto el valor que ya existe en el diccionario en una lista, y añado el nuevo valor al final.
    return diccDefinitivo

# Creo el diccionario inverso del enunciado sin valores repetidos.
diccInversoSinRepeticiones={
    "a":1,
    "b":2,
    "c":3
}

# Creo el diccionario inverso del enunciado con valores repetidos.
diccInversoConRepeticiones={
    "x":"apple",
    "y":"banana",
    "z":"banana"
}

# Llamada a la función.
diccionarioInversoValoresDuplicados(diccInversoSinRepeticiones)
diccionarioInversoValoresDuplicados(diccInversoConRepeticiones)
```

Conversió de tipus de dades

El client rep una llista de dades i necessita generar dues llistes, la primera on estaran tots els elements que es van poder convertir en flotants i l'altra on estan els elements que no es van poder convertir. Exemple de la llista que rep el client: [ '1.3', 'one' , '1e10' , 'seven', '3-1/2', ('2',1,1.4,'not-a-number'), [1,2,'3','3.4']]

```python
def listasFloatNoFloat (listaMezclada):

    # Creo las dos listas definitivas donde estarán los datos separados por tipo.
    listaFloat=[]
    listaNoFloat=[]

    # Creo una lista auxiliar para almacenar los datos internos de posibles listas,tuplas o diccionarios que puedan haber en la lista que se pasa como argumento a la función.
    # En ella también guardo los que no forman parte de ninguno de estos tipos d eestructuras de datos.
    listaAuxiliar=[] 

    # Compruebo el tipo de cada uno de los datos de la lista pasada como argumento a la función (listaMezclada) y los añado a la lista auxiliar
    # Si son una lista, tupla o diccionario añado todos sus datos mediante la función extend. En caso contrario sirve la función append.
    for auxDato in listaMezclada:
        if isinstance(auxDato,list):
            listaAuxiliar.extend(auxDato)
        elif isinstance(auxDato,tuple):
            listaAuxiliar.extend(auxDato)
        elif isinstance(auxDato,dict):
            listaAuxiliar.extend(auxDato.values())
        else:
            listaAuxiliar.append(auxDato)

    # Recorro la lista que pasan como argumento a la función.
    for dato in listaAuxiliar:
        try:                                                                      # Intento cambiar a float cada uno de los datos de la lista
            float(dato)
            listaFloat.append(float(dato))                                        # Si se puede cambiar guardo en la lista de floats
        except ValueError:                                                        # Si no se puede cambiar guardo en la lista de no floats
            listaNoFloat.append(dato)
    print("Lista de floats:", listaFloat)
    print("Lista de no floats:", listaNoFloat)

# Creo la lista
listaFloatNoFloat=['1.3', 'one' , '1e10' , 'seven', '3-1/2', ('2',1,1.4,'not-a-number'), [1,2,'3','3.4']]

# Llamada a la función.
listasFloatNoFloat(listaFloatNoFloat)
```

Comptador i endreçador de paraules d'un text.

El client va quedar content amb el comptador de paraules, però ara vol llegir arxius TXT i que calculi la freqüència de cada paraula ordenades dins de les entrades habituals del diccionari segons la lletra amb la qual comencen, és a dir, les claus han d'anar de la A a la Z i dins de la A hem d'anar de la A la Z

```python
import string

def contar_palabras_por_letra(nombreArchivoTXT):
    # Abro el archivo que paso como parámetro a la función, en modo lectura y reconociendo los carácteres especiales. Lo guardo en la variable textoLeido.
    with open(nombreArchivoTXT, 'r', encoding='utf-8') as archivoTXT:
        textoLeido=archivoTXT.read()

    # Paso el texto leído a minúsculas para que no duplique las palabras que puedan empezar por mayúsculas a principio de cada linea.
    # Limpio el texto leído ya pasado a minúsculas de caracteres especiales, incluso los no incluídos en el método string.punctuation.
    textoLeido=textoLeido.lower()
    for caracterEspecial in string.punctuation + "¡¿“”":
        textoLeido=textoLeido.replace(caracterEspecial, "")

    # Guardo en la lista palabrasSeparadas, todas las palabras del texto que he leído y separadas por un espacio en blanco.
    palabrasSeparadas=textoLeido.split()

    # Creo el diccionario repeticionesPalabras para contar las veces que se repite cada palabra.
    # Si la palabra ya existe en el diccionario, sumo una unidad más.
    # Si no está, inicio el contador a 1.
    repeticionesPalabras={}
    for palabra in palabrasSeparadas:
        if palabra in repeticionesPalabras:
            repeticionesPalabras[palabra]+=1
        else:
            repeticionesPalabras[palabra]=1

    # Creo otro diccionario palabrasAgrupadas para agrupar las palabras que empiezan por la misma letra.
    # Paso a mayúscula la primera letra de cada palabra para poder compararla con la clave del diccionario y poder clasificarla, a la vez que voy creando esas claves (A...Z).
    # Voy guardando cada palabra acorde a su primera letra.
    palabrasAgrupadas = {}
    for palabra, repeticion in repeticionesPalabras.items():
        letraInicial=palabra[0].upper()
        if letraInicial not in palabrasAgrupadas:
            palabrasAgrupadas[letraInicial]={}
        palabrasAgrupadas[letraInicial][palabra]=repeticion

    # Ordeno el diccionario agrupado con la función sorted.
    # Despues, dentro de cada clave ordeno los valores.
    # Imprimo cada clave y cada valor.
    for letraAbecedario in sorted(palabrasAgrupadas.keys()):
        print(letraAbecedario)
        for palabra in sorted(palabrasAgrupadas[letraAbecedario].keys()):
            print(f"{palabra}:{palabrasAgrupadas[letraAbecedario][palabra]}")

# copio el nombre del archivo que quiero procesar a la variable nombreArchivoTXT, que es el argumento que paso a la función.
nombreArchivoTXT=r"C:\Users\bnova\OneDrive\Escritorio\IT ACADEMY\IT ACADEMY\03_Especialización (EDA)\Sprint 07\tu_me_quieres_blanca.txt"

# Llamo a la función.
contar_palabras_por_letra(nombreArchivoTXT)
```
